    a) Create a figure object called fig using plt.figure()
    b) Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
    c) Plot (x,y) on that axes and set the labels and titles to match the plot below:



```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(0,100)
y=x*2
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.show()
```

```python
fig1=plt.figure()
ax1=fig1.add_axes([0,0,1,1])
ax2=fig1.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax2.plot(x,y)
plt.show()
```

Use the company sales dataset csv file, read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: –
    a. X label name = Month Number
    b. Y label name = Total profit


```python
import pandas as pd
```

```python
data=pd.read_csv('company_sales_data.csv')
month=data['month_number']
profit=data['total_profit']
plt.figure()
plt.plot(month,profit)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Profit of Company Sales')
plt.show()
```

Use the company sales dataset csv file, get total profit of all months and show line plot with the following Style properties. Generated line plot must include following Style properties: –
    a. Line Style dotted and Line-color should be red
    b. Show legend at the lower right location.
    c. X label name = Month Number
    d. Y label name = Sold units number
    e. Add a circle marker.
    f. Line marker color as read
    g. Line width should be 3 

```python
month=data['month_number']
units=data['total_units']
fig,ax=plt.subplots()
ax.plot(month,units,linestyle='dotted',color='red')
ax.legend(loc='lower right')
plt.xlabel('Month Number')
plt.ylabel('Total units')
plt.title('Profit of Company Sales')
plt.show()
```

1. Use the company sales dataset csv file, read all product sales data and show it  using a multiline plot.            
                       Display the number of units sold per month or each product using multiline plots. (i.e., Separate Plotline                           
                       for each product ).

```python
month = data['month_number']
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
product_sales = [data[product] for product in products]
fig, ax = plt.subplots()

for product_sale, product in zip(product_sales, products):
    ax.plot(month, product_sale, label=product)

ax.set_xlabel('Month Number')
ax.set_ylabel('Units Sold')
ax.set_title('Product Sales per Month')
ax.legend()
plt.show()
```
